```python
# Your code goes in here
numerical_var = df.select_dtypes(include='number').columns.tolist()
categorical_var = df.select_dtypes(include='object').columns.tolist()
numerical_var = [col for col in numerical_var if col != 'SalePrice']







```

## Question 2
* Determine the categorical and numerical variables. Save the numerical variables to a python list variable named `numerical_var`. Save the categorical variables to a python list variable named `categorical_var`
* `SalePrice` is the target variable that will be used to build a regression model later on. Make sure to exlude that from the list of numerical variables.

```python
# Your code goes in here
majority_missing = [col for col in df.columns if df[col].isna().mean() > 0.5]
df = df.drop(columns=majority_missing)

for col in numerical_var:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

for col in categorical_var:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

```

```python
# Your code goes in here
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numerical_var] = scaler.fit_transform(df[numerical_var])

```

## Exersice 4
* Standardize the numerical variables (exlcude the `SalePrice` as usual)

```python
# Your code goes in here
for col in categorical_var:
    if col in df.columns:
        means = df.groupby(col)['SalePrice'].mean()
        df[col + '_mean_encoded'] = df[col].map(means)

```

## Exercise 6
* Encode categorical variables using target mean encoding.
* Recall what target mean encoding: Target mean encoding (also known as mean encoding or likelihood encoding) is a technique used in machine learning for encoding categorical features. This method is particularly useful when working with high-cardinality categorical variables.

```python
# your code goes in here
df = pd.get_dummies(df, columns=[col for col in categorical_var if col in df.columns], drop_first=True)
```
