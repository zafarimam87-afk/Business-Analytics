## Question 3

* Create a function named `longest_string` that takes a list of strings and returns the longest string.
* What it takes: A list named `lst` that contains strings
* What it does: it counts the number of chrachters in each string and returns the string with the most number of characters.
* What it returns:iIt returns the string with the most number of characters.

```python
# Your code goes in here
# Manual loop method
def longest_string(lst):
    longest = ""
    for s in lst:
        if len(s) > len(longest):
            longest = s
    return longest

```

* What it takes: it takes two tuples named `point1` and `point2` with two number each. point1 can be for instance (4, 8)
What it does: It computes the euclidean_distance between point1 and point2
What it returns: It returns the euclidean_distance as a number.

```python
# Your code goes in here
import math

def euclidean_distance(point1, point2):
   x1, y1,x2,y2 = point1[0], point1[1], point2[0], point2[1]
   distance = math.sqrt((x1-x2)**2+(y1-y2)**2)
   return distance




```
