# Question 7
* Are there any variables with majority missing values (more than 50%)? If yes, identify their names and save them in a list named `majority_missing_var`
* Remove these majority missing variables if exist from the dataframe df
* For the remaining variables do the following:
> * Impute the missing values in the numerical variables with the median
> * Impute the missing values in the cotegorical variables with the mode

```python
###
### YOUR CODE HERE
###
majority_missing_var = [col for col in df.columns if df[col].isna().mean() > 0.5]
df = df.drop(columns=majority_missing_var)

for col in numerical_var:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

for col in categorical_var:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

```
