# <font color='navy'>Business Context:</font>
* Attrition refers to the rate at which members of a workforce leave the company over a specific period, often measured annually. It is a natural part of business operations but can become problematic if the rate is high.
* Attrition includes employees who leave for various reasons, such as retirement, resignation, termination, or transfer to other locations. This metric is crucial for companies to track and analyze because it impacts staffing levels, workforce planning, and overall organizational efficiency.
* Predicting employee attrition is crucial for companies because it helps manage costs and maintain operational stability. High turnover rates can lead to significant expenses related to recruiting, hiring, and training new staff. Moreover, frequent turnover can disrupt workflow, impact team dynamics, and erode knowledge within a company
* By using analytics to predict when employees might leave, companies can implement targeted retention strategies, optimize their workforce planning, and ultimately sustain a more stable and experienced workforce. This proactive approach allows businesses to maintain productivity and competitiveness in their industry.

```python
hr_data.groupby('department')['dailyrate'].mean()
```

```python
avg_by_dept = (
    hr_data.groupby('department', dropna=False)['dailyrate']
           .agg(mean='mean', n='size')
           .round({'mean': 2})
           .sort_values('mean', ascending=False)
)
avg_by_dept
```
