# Question 6
* repeat the same analysis. But this time, create seperate histograms of montlycharges for each contract type.
* Recall that there are three contract types month-to-month, One year, and Two year. So you will have three histograms
* You can use plt.subplot and for loops, or any other method.

```python
###
### YOUR CODE GOES HERE
###
contract_types = df['Contract'].unique()

plt.figure(figsize=(15, 4))
for i, contract in enumerate(contract_types):
    plt.subplot(1, 3, i + 1)
    subset = df[df['Contract'] == contract]
    plt.hist(subset['MonthlyCharges'], bins=15, edgecolor='black')
    plt.title(contract)
    plt.xlabel('Monthly Charges')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
###
### YOUR INTERPRETATION GOES HERE

#This set of histograms shows how MonthlyCharges are distributed across different contract types (Month-to-month, One year, Two year).

#Month-to-month customers show a wide spread with a higher peak at lower charges, suggesting more variation and a concentration of lower spenders.

#One year and Two year contracts tend to have slightly higher charges and more uniform distribution, which could reflect bundled or premium service plans that come with longer commitments.
```

# Question 10
* Create a heatmap summarizing MonthlyCharges by Sentiment and Contract Type

```python
###
### YOUR CODE GOES HERE
###
pivot = df.pivot_table(values='MonthlyCharges', index='Sentiment', columns='Contract', aggfunc='mean')

plt.figure(figsize=(8, 5))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Average Monthly Charges by Sentiment and Contract')
plt.ylabel('Sentiment')
plt.xlabel('Contract')
plt.show()
```

```python
###
### YOUR INTERPRETATION GOES HERE

#This heatmap shows the average MonthlyCharges for customers, broken down by their Sentiment (likely reflecting customer satisfaction or tone) and Contract Type.

#Negative sentiment customers tend to have higher monthly charges, especially on Month-to-month contracts.

#Positive sentiment customers are generally paying less on average.

#This suggests that high prices in short-term contracts may be a source of dissatisfaction, possibly influencing churn and customer complaints.
```
